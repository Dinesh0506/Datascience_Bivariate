```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def  calc_vif(x):
    #calculating vif 
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return (vif)
```

```python
calc_vif(dataset[["etest_p","salary"]])
```

```python
#function of Calc_vif()
#1.creating temple for vif values
#2.assigning values as variable to X
#3.in this first for loop range in x values followed by i.
#then variance inflation factor function calculated 
#the values instence i values in to x values finally all are saved in vif 
```

```python
#Independent Sample-Unpaired T Test
#Different Group(Male,Female) but Same Condition(Salary)
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Male=dataset[dataset["gender"]=="M"]["salary"]
Female=dataset[dataset["gender"]=="F"]["salary"]
ttest_ind(Male,Female)
```

```python
#Dependent Sample-Paired T-Test
#Same Group(Male)but Different Condition(SSC_P,HSC_P)
```

```python
from scipy.stats import ttest_rel
Male=dataset[dataset["gender"]=="M"]["ssc_p"]
Male1=dataset[dataset["gender"]=="M"]["hsc_p"]
ttest_rel(Male,Male1)
```

```python
#Independent Sample--> Unpaired T_Test
#Difference between Hsc_b Group(others,Central)and Condition(Salary)
from scipy.stats import ttest_ind
Male=dataset[dataset["hsc_b"]=="Others"]["salary"]
Female=dataset[dataset["hsc_b"]=="Central"]["salary"]
ttest_ind(Male,Female)
```

```python
#Dependent Sample--> Paired T_Test
#same Group(Commerce)and diferent condition(mba_p,etest_p)
from scipy.stats import ttest_rel
Male=dataset[dataset["hsc_s"]=="Commerce"]["mba_p"]
Male1=dataset[dataset["hsc_s"]=="Commerce"]["etest_p"]
ttest_rel(Male,Male1)
```

```python
#Dependent Sample--> Paired T_Test
#same Group(Science)and diferent condition(mba_p,etest_p)
from scipy.stats import ttest_rel
Male=dataset[dataset["hsc_s"]=="Science"]["mba_p"]
Male1=dataset[dataset["hsc_s"]=="Science"]["etest_p"]
ttest_rel(Male,Male1)
```

```python
#In this t_test for science and commerce in mba_p ,etest_p 
#is higher difference and not a acceptable range.
```

```python
# Hypothesis Test
```

```python
#Statement: test whether the SSC_P of male and HSC_P of male marks is same?
```

```python
#Null Hypothesis--> there is no  similarity between ssc_p pass mark of male and hsc_p pass mark of male
```

```python
#Dependent Sample--> Paired T_Test
#same Group(Male)and diferent condition(ssc_p,hsc_p)
from scipy.stats import ttest_rel
Male=dataset[dataset["gender"]=="M"]["ssc_p"]
Male1=dataset[dataset["gender"]=="M"]["hsc_p"]
ttest_rel(Male,Male1)
```

```python
#Solution:-
#Rejected Null Hypothesis  p<0.05
#But in this statement  pvalue is higher then 0.05.
#Then Accept the null hypothesis and rejected Alternate Hypotheis testing.
```

```python
# ANAVO --> Analysis Of Variance 
```

```python
#HO--> there is no significant difference in marks
#H1-->there is significant difference in marks
```

```python
#ONE Way Classification 
import scipy.stats as stats
stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['mba_p'])
```

```python
#Solution:-
#Rejected the Null Hypothesis p<0.05.
#Then Accept the Alternate Hypothesis
# Here the difference between ssc_p ,hsc_p and mba_p marks is higher the p value.
#Then Null Hypothesis there is no significant difference between ssc_p,hsc_p,mba_p marks.
#Accept the Null Hypothesis and Reject the Alternate Hypothesis in this statement of one way Classification.
```
