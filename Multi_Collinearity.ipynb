```python
#Calculating correlation in dataset
def correlation(df, threshold):
    correlated_cols = set()
    corr_matrix = df.corr(numeric_only=True)
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                correlated_cols.add(colname)
    return correlated_cols
```

```python
#Calling the fuction 
correlation(X_train, 0.4)
```

```python
#Assigning 
corr_feature = correlation(X_train, 0.4)
```

```python
#Removing the highly corelated columns in dataset
X_train.drop(labels=corr_feature, axis=1, inplace=True)
X_test.drop(labels=corr_feature, axis=1, inplace=True)
```
